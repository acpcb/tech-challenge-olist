```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

orders = pd.read_csv("olist_orders_dataset.csv")
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
items = pd.read_csv("olist_order_items_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
payments = pd.read_csv("olist_order_payments_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
translation = pd.read_csv("product_category_name_translation.csv")

date_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_columns:
    orders[col] = pd.to_datetime(orders[col])

# filta períodos completos
data_inicio = "2016-09-01"
data_fim = "2018-08-31"

orders = orders[
    (orders["order_purchase_timestamp"] >= data_inicio) &
    (orders["order_purchase_timestamp"] <= data_fim)
].copy()
```

```python
df_vendas = orders.merge(
    items[["order_id", "price"]],
    on="order_id",
    how="left"
)

df_vendas["ano_mes"] = df_vendas["order_purchase_timestamp"].dt.to_period("M")

evolucao_mensal = df_vendas.groupby("ano_mes").agg({
    "order_id": "nunique",
    "price": "sum"
}).rename(columns={
    "order_id": "total_pedidos",
    "price": "receita_total"
})

evolucao_mensal["receita_total"].plot(
    figsize=(12,6)
)

plt.title("Evolução Mensal da Receita")
plt.ylabel("Receita Total (R$)")
plt.xlabel("Ano/Mês")
plt.show()
```

```python
orders["delivery_delay_days"] = (
    orders["order_delivered_customer_date"] -
    orders["order_estimated_delivery_date"]
).dt.days

orders["is_delayed"] = np.where(
    orders["delivery_delay_days"] > 0,
    "Atrasado",
    "No Prazo"
)

df = orders.merge(
    reviews[["order_id", "review_score"]],
    on="order_id",
    how="left"
)

resultado = df.groupby("is_delayed")["review_score"].mean()

resultado.plot(kind="bar", figsize=(8,5))

plt.title("Impacto do Prazo de Entrega na Experiência do Cliente")
plt.ylabel("Nota Média de Avaliação")
plt.xlabel("")
plt.xticks(rotation=0)
plt.show()
```

```python
if "customer_state" not in df.columns:
    df = df.merge(
        customers[["customer_id", "customer_state"]],
        on="customer_id",
        how="left"
    )

df_atrasado = df[df["delivery_delay_days"] > 0]

atraso_estado_real = df_atrasado.groupby(
    "customer_state"
)["delivery_delay_days"].mean()

top_estados_atraso = atraso_estado_real.sort_values(
    ascending=False
).head(10)

top_estados_atraso.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Rotas Críticas: Estados com Maior Atraso Médio")
plt.ylabel("Atraso Médio em Dias")
plt.xlabel("Estado")
plt.xticks(rotation=0)
plt.show()
```

```python
df_receita = df.merge(
    items[["order_id", "price", "freight_value"]],
    on="order_id",
    how="left"
)

receita_status = df_receita.groupby("is_delayed")[
    ["price", "freight_value"]
].sum()

receita_status["price"].plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Receita por Status de Entrega")
plt.ylabel("Receita Total (R$)")
plt.xlabel("")
plt.xticks(rotation=0)
plt.show()
```

```python
payments_filtrado = payments.merge(
    orders[["order_id"]],
    on="order_id",
    how="inner"
)

pagamentos = payments_filtrado.groupby("payment_type")[
    "payment_value"
].sum().sort_values(ascending=False)

pagamentos.index = pagamentos.index.map({
    "credit_card": "Cartão de Crédito",
    "boleto": "Boleto",
    "voucher": "Voucher",
    "debit_card": "Cartão de Débito",
    "not_defined": "Outros"
})

pagamentos.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Distribuição de Receita por Forma de Pagamento")
plt.ylabel("Receita Total (R$)")
plt.xlabel("")
plt.xticks(rotation=25)
plt.show()
```

```python
products = products.merge(
    translation,
    on="product_category_name",
    how="left"
)

items_filtrado = items.merge(
    orders[["order_id"]],
    on="order_id",
    how="inner"
)

df_categoria = items_filtrado.merge(
    products[["product_id", "product_category_name_english"]],
    on="product_id",
    how="left"
)

top_cat = df_categoria.groupby(
    "product_category_name_english"
)["price"].sum().sort_values(ascending=False).head(10)

mapa_categorias = {
    "health_beauty": "Saúde e Beleza",
    "watches_gifts": "Relógios e Presentes",
    "bed_bath_table": "Cama, Mesa e Banho",
    "sports_leisure": "Esporte e Lazer",
    "computers_accessories": "Informática e Acessórios",
    "furniture_decor": "Móveis e Decoração",
    "cool_stuff": "Utilidades Gerais",
    "housewares": "Utensílios Domésticos",
    "auto": "Automotivo",
    "garden_tools": "Jardim e Ferramentas"
}

top_cat.index = top_cat.index.map(mapa_categorias)

top_cat.plot(
    kind="bar",
    figsize=(12,6)
)

plt.title("Principais Fontes de Receita por Categoria")
plt.ylabel("Receita Total (R$)")
plt.xlabel("")
plt.xticks(rotation=30)
plt.show()
```
